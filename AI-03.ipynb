### 原始：构造x和y列表画图

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
nbSamples = 256

xRange = (-math.pi, math.pi)

x, y = [], []

# x = []
# y = []

for n in range(nbSamples):
    k = (n + 0.5) / nbSamples  #compute percentage
    x.append(xRange[0] + (xRange[1] - xRange[0]) * k)
    y.append(math.sin(x[-1]))

plt.plot(x, y)
plt.show()
```

## 提升：利用NumPy构造坐标 

```python
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#Generate a sinusoid 正弦曲线
nbSamples = 256
xRange = (-math.pi, math.pi)

x, y = np.zeros(nbSamples), np.zeros(nbSamples)

for n in range(nbSamples):
    k = (n + 0.5) / nbSamples
    x[n] = xRange[0] + (xRange[1] - xRange[0]) * k
    y[n] = math.sin(x[n])

# Plot the sinuoid
plt.plot(x, y)
plt.show()
```

### 提升2：利用Numpy+ linspace构造坐标


numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

在指定的间隔内返回均匀间隔的数字。

返回num均匀分布的样本，在[start, stop]。


```python
import math
import matplotlib.pyplot as plt
import numpy as np

#Generate a sinusoid 正弦曲线
nbSamples  = 256
x = np.linspace(-math.pi, math.pi, num = 256)
y = np.zeros(nbSamples)

for n in range(nbSamples):
    y[n] = math.sin(x[n])

# Plot the sinuoid
plt.plot(x, y)
plt.show()
```

### 高级一点的图注

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')
ax.axis('equal')

# specify the lines and labels of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper right', frameon = False)

# Create the second legend and add the artist manually.
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower right', frameon = False)
ax.add_artist(leg);
```

### 设置x或y轴范围

```python
import math
import matplotlib.pyplot as plt
import numpy as np

#Generate a sinusoid and cosinoid 正弦和余弦曲线

x = np.linspace(-math.pi, math.pi, num = 256)
y = np.sin(x)
z = np.cos(x)

 #Axis setup
fig = plt.figure()
axis = fig.add_subplot(111) 

axis.set_ylim(-0.5 * math.pi, 0.5 * math.pi)    #<--Set the y-axis view limits.

plt.plot(x, y, 'r--', label= 'sin(x)')
plt.plot(x, z,'b-',linewidth = 3, label = 'cos(x)')
# legend location can be 'best' , ' center' , 'left' , 'right ' , e t c .

plt.legend(loc = 'best')
plt.show()
```

### 子图参数的意义

ax = fig.add_subplot(349)

参数349的意思是：将画布分割成3行4列，图像画在从左到右从上到下的第9块

```python
import math
import matplotlib.pyplot as plt
import numpy as np

#Generate a sinusoid and cosinoid 正弦和余弦曲线

x = np.linspace(-math.pi, math.pi, num = 256)
y = np.sin(x)
z = np.cos(x)

#Axis setup
fig = plt.figure()
axis = fig.add_subplot(211) 

axis.set_ylim(-0.5 * math.pi, 0.5 * math.pi)    #Set the y-axis view limits.
axis.grid(True)  #<--
plt.plot(x, y, 'r--', label= 'sin(x)')
plt.legend(loc = 'best')

axis = fig.add_subplot(212) #<--
plt.plot(x, z,'b-',linewidth = 3, label = 'cos(x)')
# legend location can be 'best' , ' center' , 'left' , 'right ' , e t c .

plt.legend(loc = 'best')
plt.show()

##Q:如果让图片一行两列
```

### 散点图

```python
import matplotlib.pyplot as plt
import numpy as np

#产生一个2D的正太分布
nbPointers = 50
x = np.random.standard_normal(nbPointers)
y = np.random.standard_normal(nbPointers)

# Fixing random state for reproducibility
np.random.seed(19680801)
colors = np.random.rand(nbPointers)

#plt.scatter(x, y)

area = (30 * np.random.rand(nbPointers))**2  # 0 to 15 point radii
plt.scatter(x, y, s = area, c = colors, alpha = 0.5)


plt.show()
```
